```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
df.info() 
```

```python
df.describe()
```

```python
df.columns
```

```python
sns.scatterplot(x='area', y='price', data=df, alpha=0.7)
plt.title("Area vs Price");
```

```python
sns.jointplot(x='area', y='price', data=df, palette='coolwarm')
```

```python
df.groupby('bedrooms')['price'].mean().plot(kind='bar')
plt.title("Mean price for no. of bedrooms")
```

```python
sns.boxplot(x='bathrooms', y='price', data=df)
plt.title("Bathrooms vs Price");
```

```python
sns.swarmplot(x='stories', y='price', data=df)
plt.title("Stories vs Price");
```

```python
plt.hist(df['price'], bins=20)
plt.title("Distribution of Price");
```

```python
plt.hist(df['area'], bins=20)
plt.title("Distribution of Area");
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Heatmap of Correlation of Data");
```

```python
df.groupby('mainroad')['price'].mean().plot(kind='bar')
plt.title("Mainroad vs Price");
```

```python
guestroom_counts = df['guestroom'].value_counts()

plt.figure(figsize=(6,6))  # Set the figure size
plt.pie(guestroom_counts, labels=guestroom_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99'])
plt.title('Guest Room Availability')
```

```python
sns.boxplot(x='basement', y='price', data=df)
plt.title("Basement vs. Price")
```

```python
sns.countplot(x='bedrooms', data=df)
```

```python
plt.figure(figsize=(10,6))

scatter = plt.scatter(df['area'], df['price'], 
                      c=df['bedrooms'],  
                      s=df['bathrooms']*100,  
                      cmap='viridis',    # change color map
                      alpha=0.7, edgecolor='w')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Bedrooms')

plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot: Area vs Price with Bedrooms (color) and Bathrooms (size)');
```
